# 04 — TCR Reactivity Selection

Ranks clonotypes by exhaustion score and selects the top tumor-reactive candidates.

```python
# Papermill parameter cell
# Default uses scVI outputs; CCA pipeline overrides to data/cca/ranked_clonotypes.csv
clonotype_table_path = 'data/scvi/ranked_clonotypes.csv'
model_path           = 'models/rf_tumor_reactivity.pkl'
output_selected_tcr  = 'data/scvi/selected_tumor_reactive_tcrs.csv'
```

```python
import pandas as pd
import os

print(f'clonotype_table : {clonotype_table_path}')
print(f'output          : {output_selected_tcr}')
```

```python
# Load exhaustion-ranked clonotypes from NB03
assert os.path.exists(clonotype_table_path), f'Missing: {clonotype_table_path}'
df = pd.read_csv(clonotype_table_path)
print(f'Loaded {len(df)} clonotypes, columns: {list(df.columns)}')
df.head()
```

```python
# Select top clonotypes by exhaustion score
score_col = next((c for c in df.columns if 'score' in c.lower() or 'exhaust' in c.lower()), None)
if score_col:
    selected = df.sort_values(score_col, ascending=False).head(20).copy()
    print(f'Top 20 clonotypes by {score_col}:')
    print(selected[['clonotype_id', score_col]].to_string())
else:
    selected = df.head(20).copy()
    print('No score column found; taking first 20 rows')
```

```python
# Save selected TCRs
os.makedirs(os.path.dirname(os.path.abspath(output_selected_tcr)), exist_ok=True)
selected.to_csv(output_selected_tcr, index=False)
print(f'Saved {len(selected)} selected TCRs -> {output_selected_tcr}')
```
